##### 4.- Escribe una función que cuente el porcentaje de A de una secuencia

```python
def porcentaje_A(secuencia):

    secuencia = secuencia.upper()

    total = len(secuencia)

    if total == 0:

        return 0

    num_A = secuencia.count('A')

    print("Número de A:", num_A)

    print("Longitud total:", total)

    porcentaje = (num_A / total) * 100

    return porcentaje

```

```python
dna = "ATCGATCGATCGAAATTT"

print("Porcentaje de A:", porcentaje_A(dna), "%")
```

##### ¿Puedes hacer que devuelva dos decimales? ¿Y que lidie con mayúsculas y minúsculas?

```python
def porcentaje_A(secuencia):
    secuencia = secuencia.upper().strip()  # uniformiza y limpia espacios
    total = len(secuencia)
    
    if total == 0:
        return 0.00  # evita división por cero si la secuencia está vacía
    
    count_A = secuencia.count('A')
    porcentaje = (count_A / total) * 100
    return round(porcentaje, 2)
```

```python
dna = "aTcGaTcGaAaT"
resultado = porcentaje_A(dna)
print(f"Porcentaje de A: {resultado:.2f}%")
```

##### ¿Y que devuelva el número de decimales que yo quiero?

```python
def porcentaje_A(secuencia, decimales=2):
    secuencia = secuencia.upper().strip()
    total = len(secuencia)
    
    if total == 0:
        return 0.0
    
    count_A = secuencia.count('A')
    porcentaje = (count_A / total) * 100
    return round(porcentaje, decimales)
```

```python
dna = "aTcGaTcGaAaT"

print(porcentaje_A(dna))          
print(porcentaje_A(dna, 3))       
print(porcentaje_A(dna, 0)) 
```

#### 5.- De la lista de accessions que se proporciona obtén lo siguiente:
1. Los que tengan el número 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o más dígitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
```

```python
con_5 = [a for a in accs if '5' in a]
```

```python
con_d_o_e = [a for a in accs if re.search(r"[de]", a)]
```

```python
con_de = [a for a in accs if re.search(r"d.*e", a)]
```

```python
con_d_letra_e = [a for a in accs if re.search(r"d.e", a)]
```

```python
con_d_y_e = [a for a in accs if 'd' in a and 'e' in a]
```

```python
empiezan_x_y = [a for a in accs if re.match(r"^[xy]", a)]
```

```python
empiezan_x_y_terminan_e = [a for a in accs if re.match(r"^[xy].*e$", a)]
```

```python
tres_o_mas_digitos = [a for a in accs if re.search(r"\d{3,}", a)]
```

```python
terminan_d_ar_p = [a for a in accs if re.search(r"d[arp]$", a)]
```

```python
print("1. Con 5:", con_5)
print("2. Con d o e:", con_d_o_e)
print("3. Con d y e en ese orden:", con_de)
print("4. Con d y una letra y luego e:", con_d_letra_e)
print("5. Con d y e en cualquier orden:", con_d_y_e)
print("6. Empiezan por x o y:", empiezan_x_y)
print("7. Empiezan por x o y y terminan en e:", empiezan_x_y_terminan_e)
print("8. Tres o más dígitos seguidos:", tres_o_mas_digitos)
print("9. Terminan con d seguido de a/r/p:", terminan_d_ar_p)
```
