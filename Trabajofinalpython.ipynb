#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
def digestión_ADN(secuencia, enzR):

    enzimas_corte = {}
    enzimas_no_corte = []
    
    for enzima, sitio in enzR.items():
        # Determino la posición relativa de corte en el sitio
        corte_relativo = sitio.find('*')
        sec_sin_asterisco = sitio.replace('*', '')
        
        # Busco todas las posiciones de corte en la secuencia
        posiciones = []
        inicio = 0
        while True:
            idx = secuencia.find(sec_sin_asterisco, inicio)
            if idx == -1:
                break
            posiciones.append(idx + corte_relativo)  # Ajustamos la posición con el asterisco
            inicio = idx + 1  # Continuamos la búsqueda después de este sitio
        
        if posiciones:
            enzimas_corte[enzima] = posiciones
        else:
            enzimas_no_corte.append(enzima)
    
    # Generar fragmentos de la digestión
    # cojo todas las posiciones de corte de todas las enzimas, ordenadas
    todas_posiciones = []
    for pos_list in enzimas_corte.values():
        todas_posiciones.extend(pos_list)
    todas_posiciones = sorted(todas_posiciones)
    
    # Corto la secuencia en fragmentos
    fragmentos = []
    inicio = 0
    for pos in todas_posiciones:
        fragmentos.append(secuencia[inicio:pos])
        inicio = pos
    fragmentos.append(secuencia[inicio:])  # último fragmento
    
    # Tamaño de cada fragmento
    info_fragmentos = [len(f) for f in fragmentos]
    
    # devuelvo resultados
    return {
        "enzimas_corte": enzimas_corte,
        "enzimas_no_corte": enzimas_no_corte,
        "fragmentos": fragmentos,
        "info_fragmentos": info_fragmentos
    }

```

```python
# Ejemplo
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}

resultado = digestión_ADN(sequence, enzR)
print("Enzimas que cortan y posiciones de corte:\n", resultado["enzimas_corte"])
print("\nEnzimas que no cortan:\n", resultado["enzimas_no_corte"])
print("\nFragmentos resultantes:\n", resultado["fragmentos"])
print("\nNúmero y tamaño de cada fragmento:\n", resultado["info_fragmentos"])
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python

codon_table = {
    'ATA':'I','ATC':'I','ATT':'I','ATG':'M',
    'ACA':'T','ACC':'T','ACG':'T','ACT':'T',
    'AAC':'N','AAT':'N','AAA':'K','AAG':'K',
    'AGC':'S','AGT':'S','AGA':'R','AGG':'R',
    'CTA':'L','CTC':'L','CTG':'L','CTT':'L',
    'CCA':'P','CCC':'P','CCG':'P','CCT':'P',
    'CAC':'H','CAT':'H','CAA':'Q','CAG':'Q',
    'CGA':'R','CGC':'R','CGG':'R','CGT':'R',
    'GTA':'V','GTC':'V','GTG':'V','GTT':'V',
    'GCA':'A','GCC':'A','GCG':'A','GCT':'A',
    'GAC':'D','GAT':'D','GAA':'E','GAG':'E',
    'GGA':'G','GGC':'G','GGG':'G','GGT':'G',
    'TCA':'S','TCC':'S','TCG':'S','TCT':'S',
    'TTC':'F','TTT':'F','TTA':'L','TTG':'L',
    'TAC':'Y','TAT':'Y','TAA':'*','TAG':'*',
    'TGC':'C','TGT':'C','TGA':'*','TGG':'W',
}

def revcomp(seq):
    comp = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    return "".join(comp[b] for b in seq[::-1])

def translate(seq):
    prot = ""
    for i in range(0, len(seq), 3):
        cod = seq[i:i+3]
        if len(cod) < 3:
            break
        aa = codon_table.get(cod, 'X')
        if aa == "*":   # Stop codon → terminamos la proteína
            break
        prot += aa
    return prot

def find_orfs(seq, strand):
    orfs = []
    stops = {"TAA", "TAG", "TGA"}
    
    for frame in range(3):
        i = frame
        while i < len(seq)-2:
            codon = seq[i:i+3]
            if codon == "ATG":  # Inicio de ORF
                j = i
                while j < len(seq)-2:
                    cod2 = seq[j:j+3]
                    if cod2 in stops:  # Fin del ORF
                        dna = seq[i:j+3]
                        orfs.append({
                            "start": i,
                            "end": j+3,
                            "frame": frame,
                            "strand": strand,
                            "dna": dna,
                            "protein": translate(dna)
                        })
                        break
                    j += 3
                i = j
            i += 3
    return orfs


def analizar_ORFs(sequence):

    # 1) ORFs de ambas hebras
    orfs_pos = find_orfs(sequence, "+")
    orfs_neg = find_orfs(revcomp(sequence), "-")
    
    all_orfs = orfs_pos + orfs_neg

    # 2) Ordenar por longitud
    all_orfs_sorted = sorted(all_orfs, key=lambda x: len(x["protein"]), reverse=True)

    # 6 ORFs más largos
    top6 = all_orfs_sorted[:6]

    # Mejor predicción = ORF más largo
    best = top6[0]

    # 3) Porcentaje aromáticos
    arom = sum(aa in "FWY" for aa in best["protein"])
    pct_arom = (arom / len(best["protein"])) * 100 if best["protein"] else 0

    # 4) Longitud relativa al ADN
    long_rel = len(best["dna"]) / len(sequence) * 100

    return {
        "total_orfs": len(all_orfs),
        "top6": top6,
        "best": best,
        "pct_aromaticos": pct_arom,
        "longitud_relativa": long_rel
    }
```

```python
with open("sequence.fasta", "r") as f:
    lines = f.readlines()

sequence = "".join(line.strip() for line in lines[1:])
```

```python
resultado = analizar_ORFs(sequence)

print("Total ORFs:", resultado["total_orfs"])
print("Mejor ORF:", resultado["best"])
print("Porcentaje aromáticos:", resultado["pct_aromaticos"])
print("Longitud relativa:", resultado["longitud_relativa"])
```
