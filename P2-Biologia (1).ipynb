#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
#eliminar duplicados
```

```python
dinucleotides = list(set(dinucleotides))
```

```python
#diccionario
```

```python
conteos = {}
```

```python
for di in dinucleotides:
    conteos[di] = dna.count(di)
```

```python
for di, count in conteos.items():
    print(f"{di}: {count}")
```

##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
conteos = {}
for di in set(dinucleotides):  # usamos set para eliminar duplicados
    conteos[di] = dna.count(di)
```

```python
print("Dinucleótidos con count = 2:")
for di, count in conteos.items():
    if count == 2:
        print(f"{di}: {count}")
```
